```python
# ============= Tensoren erstellen

# Nach der Größe 
w = torch.empty(100,200) 
w = torch.zeros(100,200) 
w = torch.ones(100,200)

# Nach Größe mit Zufallswerten 
w = torch.rand(100,200) 
w = torch.randn(100,200) 
w = torch.randint(5,10,(100,200)) # (low, high (size=tuple defining the shape))

# Mit angegebenem Datentyp oder Gerät initialisiert.
w = torch.empty((100,200), 
                dtype=torch.float64,
                device="cpu") #devide="cuda"

# Mit gleicher Größe, gleichem Datentyp und gleichem Gerät wie ein anderer Tensor initialisiert.
x = torch.empty_like(w)
```

```python
# ====== Tensorattribute

A = torch.tensor([[2,3],
                [4,5]])

print(".dtype: ", A.dtype)
print (".layout: ", A.layout)
print(".device: ", A.device)
print(".shape: ", A.shape)  # .size() und .shape ist das selbe!
print(".size(): ", A.size())
print(".ndim: ", A.ndim)
```

```python
# ==== Tensor indizieren
B = torch.tensor([[1,2],
                [3,4],
                [5,6]])
print(B)

# Indizierung: Gibt einen Tensor zurück!
print(B[0,1])

# Indizierung: gibt einen Wert als Python Zahl zurück
print(B[0,1].item())

# ===== Slicing
print (B[:2,0]) # :2 - EXKLUSIVE 2!

# ===== Boolesches Indizieren
print(B[B<5])

# === View (shares memory with original tensor!)

```

```python
# --- Iterate through Tensor
# 1D tensor: Vector

v = torch.randint(0,30,(3,))
print(v)
print(v.size())

v1 = torch.tensor([[1,2,3],
                  [4,5,6]])
print(v1.size())
print(v1.size(0))
print(v1.size(1))
# print(v1.shape(0)) doesnt work!
```

```python
A = torch.Tensor([[2,3],
                [4,5]])
a_storage = A.storage()
print("Storage:\n", a_storage)
print("Shape:\n ", A.shape)
print("View:\n ", A.view(1,4))
```

Storage = one-dimensional array of numerical data: a contiguous block of memory containing numbers of a given type
Tensor = view on storage

Offset = index in the storage corresponding to the first element in the tensor
Size/shape = tuple indicating how many elements across each dimension the tensor represents
Stride = the number of elements in the storage that need to be skipped over to
obtain the next element along each dimension

calling view on a tensor returns a new ten-
sor that changes the number of dimensions and the striding information, without
changing the storage.
