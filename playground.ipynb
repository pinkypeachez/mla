**torch.permute(input, dims)**

Parameter:
- input (Tensor) – the input tensor.
- dims (torch.Size, tuple of int or list of int) – the desired ordering of dimensions.

**torch.reshape(input, dims)**

*Returns a tensor with the same data and number of elements as input, but with the specified shape.*

*When possible, the returned tensor will be a view of input. Otherwise, it will be a copy.*

Parameter:
- input (Tensor) – the tensor to be reshaped
- shape (tuple of int) – the new shape

```python
# RESHAPE: DATA DOESNT MOVE IN MEMORY!
# PERMUTE: DATA MOVES IN MEMORY

#https://pytorch-types.readthedocs.io/en/latest/

# A single dimension may be -1
b = torch.tensor([[0, 1], [2, 3],[4,5]])
b2 = torch.reshape(b, (-1,)) # to single dimension: -1
print(b.shape)
print(b.cpu(memory_format=torch.preserve_format) ) #Returns a copy of this object in CPU memory.

c = torch.permute(b, (1,0))
print(c)
print(c.cpu(memory_format=torch.preserve_format) )

print(b2.shape)
print(b2.cpu(memory_format=torch.preserve_format) )
print(b2.is_contiguous())

# .is_contiguous
# Returns True if self tensor is contiguous in memory 
# in the order specified by memory format
```

```python
# VIEW
#  zero-copy Operation —  zeigt auf denselben Speicher
# Returns a new tensor with the same data as the self tensor but of a different shape

# shares the same data and must have the same number of elements!

# reshape returns a view!
b = torch.tensor([[0, 1], [2, 3],[4,5]])
print(b)
print("Shape: ", b.shape) 
print("Stride: ", b.stride()) 
b2 = b.view(6)
print("\nView: flaches Array - b.view(6)")
print(b2)
print("Stride", b2.stride())


```

```python
A = torch.Tensor([[2,3],
                [4,5]])
a_storage = A.storage()
print("Storage:\n", a_storage)
print("Shape:\n ", A.shape)
print("View:\n ", A.view(1,4))
```

Storage = one-dimensional array of numerical data: a contiguous block of memory containing numbers of a given type
Tensor = view on storage

Offset = index in the storage corresponding to the first element in the tensor
Size/shape = tuple indicating how many elements across each dimension the tensor represents
Stride = the number of elements in the storage that need to be skipped over to
obtain the next element along each dimension

calling view on a tensor returns a new ten-
sor that changes the number of dimensions and the striding information, without
changing the storage.
