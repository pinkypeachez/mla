```python
def next_power_of_2(n):
    if n <= 0: return 1
    # n.bit_length() gibt an, wie viele Bits man braucht
    return 1 << (n - 1).bit_length()

print( next_power_of_2(74))

zahl = 15
zahl.bit_length()
```
